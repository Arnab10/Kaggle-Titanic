```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
url = "https://gist.githubusercontent.com/michhar/2dfd2de0d4f8727f873422c5d959fff5/raw/ff414a1bcfcba32481e4d4e8db578e55872a2ca1/titanic.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df = pd.read_csv(url, sep="\t")
df.head()
```

```python
col_features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']
col_target = ['Survived']

X = df[col_features]
Y = df[col_target]

X.head()
Y.head()
```

```python
X.Pclass.isnull().sum()
X.Sex.isnull().sum()
X.Age.isnull().sum() # 30 NaN values
X.SibSp.isnull().sum()
X.Parch.isnull().sum()
X.Embarked.isnull().sum() # 1 NaN value
```

```python
X.Sex.replace(['male', 'female'], [1, 0], inplace=True)
X.Embarked.replace(['S', 'C', 'Q'], [1, 2, 3], inplace=True)
```

```python
X.Age.fillna(X.Age.mean(), inplace=True)
X.Embarked.fillna(2, inplace=True)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)
```

```python
clf = SVC(gamma='auto')
```

```python
clf.fit(X_train, Y_train)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test, Y_test)
```

```python
clf = tree.DecisionTreeClassifier(criterion = 'entropy')
clf.fit(X_train, Y_train)
clf.predict(X_test)
clf.score(X_test, Y_test)
```
